```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
data_merge = pd.merge(ride_data, city_data, on = "city", how = "left")

# Display the data table for preview
data_merge.head(5)
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types: urban, suburban, rural
urban_df = data_merge.loc[data_merge["type"] == "Urban"]
urban_x = urban_df.groupby(["city"]).count()["ride_id"]
urban_y = urban_df.groupby(["city"]).mean()["fare"]
urban_d = urban_df.groupby(["city"]).mean()["driver_count"]

suburban_df = data_merge.loc[data_merge["type"] == "Suburban"]
suburban_x = suburban_df.groupby(["city"]).count()["ride_id"]
suburban_y = suburban_df.groupby(["city"]).mean()["fare"]
suburban_d = suburban_df.groupby(["city"]).mean()["driver_count"]

rural_df = data_merge.loc[data_merge["type"] == "Rural"]
rural_x = rural_df.groupby(["city"]).count()["ride_id"]
rural_y = rural_df.groupby(["city"]).mean()["fare"]
rural_d = rural_df.groupby(["city"]).mean()["driver_count"]

# Build the scatter plots for each city types
#alpha is transparency overlap
#s is multiplied by 10 to scale
plt.scatter(urban_x, urban_y, s = urban_d * 10, label = "Urban", color = ["Coral"], edgecolor = "black", alpha = .7, linewidths = 1)
plt.scatter(suburban_x, suburban_y, s = suburban_d * 10, label = "Suburban", color = ["Skyblue"], edgecolor = "black", alpha = .7, linewidths = 1)
plt.scatter(rural_x, rural_y, s = rural_d * 10, label = "Rural", color = ["Gold"], edgecolor = "black", alpha = .7, linewidths = 1)
plt.grid()

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

# Create a legend
legend = plt.legend(title = "City Types")
legend.legendHandles[0]._sizes = [40]
legend.legendHandles[1]._sizes = [40]
legend.legendHandles[2]._sizes = [40]

# Incorporate a text label regarding circle size
#plt.text starts at (xcoordinate, ycoordinate, "text")
plt.text(42, 35, "Note: \nCircle size correlates with driver count per city")

# Save Figure
#jpg is smaller than png
plt.savefig("Charts/PyberBubblePlot.jpg")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_percent = data_merge.groupby(["type"])["fare"].sum()

# Build Pie Chart
#autopct will place label inside wedge (%(location of text).(rounding)(f"%")%)
#explode is offset (rural_value, suburban_value, urban_value) where _value indicates distance of offset
plt.pie(type_percent, labels = ["Rural", "Suburban", "Urban"], colors = ["Gold", "Skyblue", "Coral"], shadow = True, startangle = 150, explode = (0, 0, 0.1), autopct="%.1f%%")
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("Charts/TotalFaresbyCityType.jpg")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ride_percent = data_merge.groupby(["type"])["ride_id"].count()

# Build Pie Chart
plt.pie(ride_percent, labels = ["Rural", "Suburban", "Urban"], colors = ["Gold", "Skyblue", "Coral"], shadow = True, startangle = 150, explode = (0, 0, 0.1), autopct="%.1f%%")
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("Charts/TotalRidesbyCityType")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
total_drivers = city_data.groupby(["type"])["driver_count"].sum()

# Build Pie Charts
plt.pie(total_drivers, labels = ["Rural", "Suburban", "Urban"], colors = ["Gold", "Skyblue", "Coral"], shadow = True, startangle = 160, explode = (0, 0, 0.1), autopct="%.1f%%")
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("Charts/TotalDriversbyCityType")

# Show Figure
plt.show()
```

# Observable Trends

1. Urban cities have the highest number of rides followed by suburban then rural. This can be expected due to the population density as urban cities have more people and thus require more rides. 

2. There is an inverse proportion for rides and fares. Urban cities have more number of rides and cheaper fare while rural cities have less number of rides and more expensive fare. This shows that more quantity of drivers will lead to reduced fairs. 

3. Although price fair in rural areas can be double that in urban areas, most profit is accounted for in urban cities. Urban fares is double suburban and almost ten times that of rural. Typically a driver will make more money in an urban city despite the average fare being far less as often times in rural cities drivers will not even find riders.
